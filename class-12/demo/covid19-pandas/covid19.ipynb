## List of coutries where Total Deaths > 40000 sorted by Total Deaths. Show only country name and total daeths and total cases columns?

```python
# Data where total deaths > 40000
biggest_deaths = df[ df['Total Deaths'] > 40000 ]
```

```python
biggest_deaths.describe()
```

```python
sorted_by_total_deaths = biggest_deaths.sort_values(by='Total Deaths', ascending=False)
```

```python
sorted_by_total_deaths
```

```python
final_result = sorted_by_total_deaths[['Country','Total Deaths','Total Cases']]
```
