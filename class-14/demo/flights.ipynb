# Matplotlib and Seaborn Visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([4,6,-2,7,-8,1])
```

```python
def plot(y1, y2):
    plt.figure() # Space to plot
    plt.plot(y1, label='data from y1')
    plt.plot(y2, label='data from y2')
    # To show the labels
    plt.legend()
    plt.title('Demo Seaborn')
    plt.xlabel('Default index')
    plt.ylabel('Some mock data')
    plt.show() # Show the plot
```

```python
plot([2,-7,3,9,3,-1],[-2,3,6,-9,5,1])
```

## Seaborn

```python
import seaborn as sns
```

```python
sns.set() # Set default plotter to be seaborn
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset('flights')
```

```python
data.head(30)
```

```python
sns.set_theme(style="dark")

# Plot miles per gallon against horsepower with other semantics
sns.relplot(x="year", y="passengers", hue="month", data=data)
```

```python
sns.set_theme(style="white")
sns.relplot(x='month', y='passengers', hue="year", data=data)
```

```python
grouped_by_month = data.groupby('month').sum()
sns.relplot(x='month', y='passengers', data=grouped_by_month)
```

```python
grouped_by_month
```

```python
sns.relplot(x='month', y='passengers',alpha=.5, palette="muted",
            height=6, data=grouped_by_month)
```
